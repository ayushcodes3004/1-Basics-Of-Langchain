#### Message History
We use message history class to make our model stateful


```python
from langchain_community.chat_message_histories import ChatMessageHistory # importing the ChatMessageHistory class to store the chat history
from langchain_core.chat_history import BaseChatMessageHistory #importing the base class for type hinting
from langchain_core.runnables.history import RunnableWithMessageHistory # importing the RunnableWithMessageHistory class to create a custom runnable that can access the chat history

store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model, get_session_history) # creating an instance of the RunnableWithMessageHistory class and passing the model and the get_message_history function to it
```

```python
config = {"configurable": {"session_id": "chat1"}} # creating a config dictionary with a session_id key
```

```python
response=with_message_history.invoke(
    [HumanMessage(content="Hi , My name is Ayush and I am an AI Engineer")],
    config=config
)
```

```python
response.content
```

```python
with_message_history.invoke(
    [HumanMessage(content="What is my name and what do I do?")],
    config=config
)   
```

```python
#change the configurable session_id to chat2 to start a new conversation
config = {"configurable": {"session_id": "chat2"}}
with_message_history.invoke(
    [HumanMessage(content="What is my name and what do I do?")],
    config=config
)
```

#### Prompt Template
Prompt Templates help to turn raw user information into a format that the LLM can work with.    

```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant.Amnswer all the question to the best of your ability and my name is {name}"),
        MessagesPlaceholder(variable_name="messages")
    ]
)

chain=prompt|model

chain.invoke({"name":"Ayush", "messages":[HumanMessage(content="What is my name?")]})
```

```python
chain.invoke({"name":"Ayush", "messages":[HumanMessage(content="What is my name?")]})
```

```python
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages", # Tell history wrapper where new messages are in dict input
 )  
config = {"configurable": {"session_id": "chat3"}} # creating a config dictionary with a session_id key

response = with_message_history.invoke(
    {
        "name": "Ayush",
        "messages": [HumanMessage(content="Hi, my name is Ayush and I am an AI Engineer")],
    },
    config=config,
 )
```

```python
response.content
```

### Managing the Conversation History
If left unmanaged, the list of messages will grow unbounded and potentially overflow the context window of the LLM. 


```python
from langchain_core.messages import SystemMessage, trim_messages
trimmer= trim_messages(
    max_tokens=45,
    strategy="last",
    token_counter=model,
    include_system=True,
    allow_partial=False,
    start_on="human"
)
messages = [
    SystemMessage(content="you're a good assistant"),
    HumanMessage(content="hi! I'm bob"),
    AIMessage(content="hi!"),
    HumanMessage(content="I like vanilla ice cream"),
    AIMessage(content="nice"),
    HumanMessage(content="whats 2 + 2"),
    AIMessage(content="4"),
    HumanMessage(content="thanks"),
    AIMessage(content="no problem!"),
    HumanMessage(content="having fun?"),
    AIMessage(content="yes!"),
]
trimmer.invoke(messages)
```

```python
from operator import itemgetter

from langchain_core.runnables import RunnablePassthrough

chain=(
    RunnablePassthrough.assign(messages=itemgetter("messages")|trimmer)
    | prompt
    | model
    
)

response=chain.invoke(
    {
    "name":"Ayush",
    "messages":messages + [HumanMessage(content="What ice cream do i like")],
    "language":"English"
    }
)
response.content
```

```python
response = chain.invoke(
    {
        "name": "Ayush",
        "messages": messages + [HumanMessage(content="what math problem did i ask")],
        "language": "English",
    }
)
response.content
```

```python
## Lets wrap this in the MEssage History
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages",
)
config={"configurable":{"session_id":"chat5"}}
```

```python
response=with_message_history.invoke(
    {
        "name": "Ayush",
        "messages": messages + [HumanMessage(content="what math problem did i ask")],
        "language": "English",
    },
    config={"configurable": {"session_id": "chat4"}},
)
```
